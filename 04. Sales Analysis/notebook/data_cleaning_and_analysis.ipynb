### QUESTION 1: Loss-Making Products in Tables & Bookcases

What it does: Identifies the top 20 products causing losses and provides actionable discontinuation recommendations.

```python
# Filter for problematic sub-categories
furniture_problem = sales[sales['sub_category'].isin(['Tables', 'Bookcases'])].copy()


# Aggregate profit by product
product_analysis = furniture_problem.groupby(
    ['product_id', 'product_name', 'sub_category']
).agg({
    'profit': 'sum',
    'sales': 'sum',
    'quantity': 'sum',
    'order_id': 'nunique',
    'discount': 'mean'
}).reset_index()

category_analysis = furniture_problem.groupby(['sub_category']).agg({
    'profit': 'sum',
    'sales': 'sum',
    'quantity': 'sum',
    'order_id': 'nunique',
    'discount': 'mean'}).reset_index()

display(category_analysis)
```

```python
# Calculate profit margin
product_analysis['profit_margin_%'] = (
    product_analysis['profit'] / product_analysis['sales'] * 100
).round(2)

# Sort by worst performers
product_analysis.sort_values('profit', inplace=True)

top_20_losers = product_analysis.head(20) # Top 20 loss-making products
top_20_losers.to_csv('../data/top_20_loss_making_products.csv', index=False)
```

```python
# Calculate average discounts on top 20 loss-making products
top_20_losers.groupby(['sub_category']).agg({'discount':'mean', 'profit':'mean'})
```

### **QUESTION 2: Optimal Discount Thresholds by Category**
What it does: Determines the exact discount percentage where each category becomes unprofitable.

```python
# Create discount bins for granular analysis
sales['discount_bin'] = pd.cut(
    sales['discount'],
    bins=[-0.01, 0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.40, 0.50, 1.0],
    labels=['0%', '1-5%', '6-10%', '11-15%', '16-20%',
            '21-25%', '26-30%', '31-40%', '41-50%', '50%+']
)

# Analyze by category and discount level
discount_analysis = sales.groupby(['category', 'discount_bin'], observed=True).agg({
    'profit': 'sum',
    'sales': 'sum',
    'order_id': 'count'
}).reset_index()

discount_analysis['profit_margin_%'] = (
    discount_analysis['profit'] / discount_analysis['sales'] * 100
).round(2)

# Display by category
for category in ['Furniture', 'Office Supplies', 'Technology']:
    print(f"\n{category}:")
    cat_data = discount_analysis[discount_analysis['category'] == category]
    print(cat_data[['discount_bin', 'profit_margin_%', 'sales', 'profit']])

    # Find breakpoint
    negative_point = cat_data[cat_data['profit_margin_%'] < 0]
    if len(negative_point) > 0:
        first_negative = negative_point.iloc[0]
        print(f"\nProfit turns NEGATIVE at: {first_negative['discount_bin']}")

    # Recommended max
    positive = cat_data[cat_data['profit_margin_%'] > 0]
    if len(positive) > 0:
        last_positive = positive.iloc[-1]
        print(f"\n✓ RECOMMENDED MAX DISCOUNT: {last_positive['discount_bin']}")
```

```python
# More precise analysis using exact discount values
print("EXACT BREAKPOINTS:")
for category in sales['category'].unique():
    cat_data = sales[sales['category'] == category]

    discount_profit = cat_data.groupby('discount').agg(dict(profit='sum', sales='sum')).reset_index()

    discount_profit['margin_%'] = (
        discount_profit['profit'] / discount_profit['sales'] * 100
    ).round(2)

    positive = discount_profit[discount_profit['margin_%'] > 0]
    negative = discount_profit[discount_profit['margin_%'] < 0]

    if len(positive) > 0 and len(negative) > 0:
        max_safe = positive['discount'].max()
        min_loss = negative['discount'].min()
        print(f"{category}: Safe up to {max_safe:.0%}, losses start at {min_loss:.0%}")

discount_analysis.to_csv('../data/discount_threshold_analysis.csv', index=False)
```

### QUESTION 3: Customer Segment Discount Abuse

What it does: Identifies customers who repeatedly demand high discounts and generate losses.

```python
# Segment-level discount patterns
segment_discount = sales.groupby(['segment', 'discount_bin'], observed=True).agg({
    'profit': 'sum',
    'sales': 'sum',
    'order_id': 'count'
}).reset_index()

segment_discount['profit_margin_%'] = (
    segment_discount['profit'] / segment_discount['sales'] * 100
).round(2)

print("DISCOUNT PATTERNS BY SEGMENT:")
for segment in sales['segment'].unique():
    print(f"\n{segment}:")
    seg_data = segment_discount[segment_discount['segment'] == segment]
    display(seg_data[['discount_bin', 'profit_margin_%', 'sales', 'profit']])
```

```python
# Customer-level analysis
customer_analysis = sales.groupby(['customer_id', 'customer_name', 'segment']).agg({
    'profit': 'sum',
    'sales': 'sum',
    'order_id': 'nunique',
    'discount': 'mean',
    'quantity': 'sum'
}).reset_index()

customer_analysis['profit_margin_%'] = (
    customer_analysis['profit'] / customer_analysis['sales'] * 100
).round(2)

# Flag problematic: high discount (>20%) AND negative profit
problematic = customer_analysis[
    (customer_analysis['discount'] > 0.20) &
    (customer_analysis['profit'] < 0)
].sort_values('profit')

print(f"\n\nPROBLEMATIC CUSTOMERS (High Discount + Losses):")
display(f"Total: {len(problematic)}")
display(f"Total losses: ${problematic['profit'].sum():,.2f}")
display(f"Total sales: ${problematic['sales'].sum():,.2f}")
print("\nTop 30:")
display(problematic.head(30))
```

```python
# Segment breakdown
print("\n\nBy Segment:")
prob_by_segment = problematic.groupby('segment').agg({
    'customer_id': 'count',
    'profit': 'sum',
    'sales': 'sum',
    'discount': 'mean'
})
display(prob_by_segment)
```

```python
# Identify repeat high-discount offenders
high_discount_repeat = sales[sales['discount'] >= 0.30].groupby(
    ['customer_id', 'customer_name', 'segment']
).agg({
    'order_id': 'nunique',
    'profit': 'sum',
    'sales': 'sum',
    'discount': 'mean'
}).reset_index()

high_discount_repeat = high_discount_repeat[
    high_discount_repeat['order_id'] >= 3
].sort_values('profit')

print(f"\n\nREPEAT HIGH-DISCOUNT OFFENDERS (≥30% discount, ≥3 orders):")
display(f"Count: {len(high_discount_repeat)}")
display(f"Total impact: ${high_discount_repeat['profit'].sum():,.2f}")
display(high_discount_repeat.head(20))
```

```python
# Compare discount usage across segments
segment_summary = sales.groupby('segment').agg({
    'discount': ['mean', 'median'],
    'profit': 'sum',
    'sales': 'sum'
})
print("\n\nSegment Discount Usage:")
display(segment_summary)

# Save outputs
problematic.to_csv('../data/problematic_customers_high_discount.csv', index=False)
customer_analysis.to_csv('../data/full_customer_analysis.csv', index=False)
```


### **QUESTION 4: Central Region Underperformance**

What it does: Compares Central to other regions across multiple dimensions to identify root causes.

```python
# Overall regional comparison
region_metrics = sales.groupby('region').agg({
    'sales': 'sum',
    'profit': 'sum',
    'order_id': 'count',
    'discount': 'mean',
    'quantity': 'sum'
})
region_metrics['profit_margin_%'] = (
    region_metrics['profit'] / region_metrics['sales'] * 100
).round(2)
region_metrics['avg_profit_per_order'] = (
    region_metrics['profit'] / region_metrics['order_id']
).round(2)

print("REGIONAL PERFORMANCE COMPARISON:")
display(region_metrics)
```

```python
# Category mix by region
category_region = sales.groupby(['region', 'category']).agg({
    'sales': 'sum',
    'profit': 'sum',
    'order_id': 'count'
}).reset_index()
category_region['profit_margin_%'] = (
    category_region['profit'] / category_region['sales'] * 100
).round(2)

print("\n\nCATEGORY MIX BY REGION:")
for region in sales['region'].unique():
    print(f"\n{region}:")
    region_data = category_region[category_region['region'] == region].copy()
    region_data['pct_sales'] = (
        region_data['sales'] / region_data['sales'].sum() * 100
    ).round(2)
    display(region_data[['category', 'profit_margin_%', 'pct_sales']])
```

```python
# Sub-category performance in Central
subcat_region = sales.groupby(['region', 'sub_category']).agg({
    'sales': 'sum',
    'profit': 'sum'
}).reset_index()
subcat_region['profit_margin_%'] = (
    subcat_region['profit'] / subcat_region['sales'] * 100
).round(2)

central_subcat = subcat_region[
    subcat_region['region'] == 'Central'
].sort_values('profit')

print("\n\nCENTRAL REGION - WORST SUB-CATEGORIES:")
display(central_subcat.head(10))
```

```python
# Discount analysis by region
print("\n\nDISCOUNT PATTERNS BY REGION:")
discount_region = sales.groupby(['region', 'discount_bin'], observed=True).agg({
    'profit': 'sum',
    'sales': 'sum',
    'order_id': 'count'
}).reset_index()
discount_region['profit_margin_%'] = (
    discount_region['profit'] / discount_region['sales'] * 100
).round(2)

for region in ['Central', 'West', 'East', 'South']:
    print(f"\n{region}:")
    region_data = discount_region[discount_region['region'] == region]
    display(region_data[['discount_bin', 'profit_margin_%', 'order_id']])
```

```python
# Segment mix by region
segment_region = sales.groupby(['region', 'segment']).agg({
    'sales': 'sum',
    'profit': 'sum'
}).reset_index()
segment_region['profit_margin_%'] = (
    segment_region['profit'] / segment_region['sales'] * 100
).round(2)

print("\n\nSEGMENT MIX:")
for region in sales['region'].unique():
    print(f"\n{region}:")
    display(segment_region[segment_region['region'] == region])
```

```python
# Specific problem products in Central
central_sales = sales[sales['region'] == 'Central']
central_problems = central_sales.groupby(['sub_category', 'product_name']).agg({
    'profit': 'sum',
    'sales': 'sum',
    'discount': 'mean'
}).reset_index()
central_problems['profit_margin_%'] = (
    central_problems['profit'] / central_problems['sales'] * 100
).round(2)
central_problems = central_problems[
    central_problems['profit'] < 0
].sort_values('profit')

print("\n\nTOP LOSS-MAKING PRODUCTS IN CENTRAL:")
display(central_problems.head(15))
```


### **QUESTION 5: Customer Lifetime Value Problems**

What it does: Identifies unprofitable repeat customers who should be deprioritized.

```python
# Calculate comprehensive CLV metrics
customer_clv = sales.groupby(['customer_id', 'customer_name', 'segment', 'region']).agg({
    'profit': 'sum',
    'sales': 'sum',
    'order_id': 'nunique',
    'discount': 'mean',
    'quantity': 'sum',
    'order_date': ['min', 'max']
}).reset_index()

# Flatten columns
customer_clv.columns = [
    'customer_id', 'customer_name', 'segment', 'region',
    'total_profit', 'total_sales', 'num_orders', 'avg_discount',
    'total_quantity', 'first_order', 'last_order'
]
```

```python
# Calculate additional metrics
customer_clv['profit_margin_%'] = (
    customer_clv['total_profit'] / customer_clv['total_sales'] * 100
).round(2)
customer_clv['avg_order_value'] = (
    customer_clv['total_sales'] / customer_clv['num_orders']
).round(2)
customer_clv['customer_lifespan_days'] = (
    customer_clv['last_order'] - customer_clv['first_order']
).dt.days
```

```python
# Focus on repeat customers
repeat_customers = customer_clv[customer_clv['num_orders'] >= 3].copy()
unprofitable_repeat = repeat_customers[
    repeat_customers['total_profit'] < 0
].sort_values('total_profit')

print("CLV INSIGHTS:")
display(f"Total customers: {len(customer_clv)}")
display(f"Repeat customers (≥3 orders): {len(repeat_customers)}")
display(f"Unprofitable repeat customers: {len(unprofitable_repeat)}")
display(f"Losses from unprofitable repeat: ${unprofitable_repeat['total_profit'].sum():,.2f}")

print("\n\nTOP 30 UNPROFITABLE REPEAT CUSTOMERS:")
display(unprofitable_repeat.head(30)[[
    'customer_name', 'segment', 'region', 'num_orders',
    'total_profit', 'total_sales', 'avg_discount', 'profit_margin_%'
]])
```

```python
# Segment breakdown
print("\n\nBy Segment:")
display(unprofitable_repeat.groupby('segment').agg({
    'customer_id': 'count',
    'total_profit': 'sum',
    'total_sales': 'sum'
}))

# Regional breakdown
print("\n\nBy Region:")
display(unprofitable_repeat.groupby('region').agg({
    'customer_id': 'count',
    'total_profit': 'sum'
}))
```

```python
# Order frequency analysis
print("\n\nORDER FREQUENCY VS PROFITABILITY:")
frequency_profit = repeat_customers.groupby('num_orders').agg({
    'customer_id': 'count',
    'total_profit': 'mean',
    'avg_discount': 'mean'
}).round(2)
display(frequency_profit.head(15))
```

```python
# Discount dependency
high_discount_customers = customer_clv[customer_clv['avg_discount'] >= 0.25]
print(f"\n\nHIGH DISCOUNT CUSTOMERS (≥25%):")
print(f"Total: {len(high_discount_customers)}")
print(f"Total profit: ${high_discount_customers['total_profit'].sum():,.2f}")
print(f"Profitable: {len(high_discount_customers[high_discount_customers['total_profit'] > 0])}")
print(f"Unprofitable: {len(high_discount_customers[high_discount_customers['total_profit'] < 0])}")
```

```python
# Tenure analysis
customer_clv['tenure_bucket'] = pd.cut(
    customer_clv['customer_lifespan_days'],
    bins=[-1, 0, 180, 365, 730, 9999],
    labels=['One-time', '0-6mo', '6-12mo', '1-2yr', '2yr+']
)

tenure_analysis = customer_clv.groupby('tenure_bucket', observed=True).agg({
    'customer_id': 'count',
    'total_profit': 'mean',
    'num_orders': 'mean',
    'avg_discount': 'mean'
}).round(2)

print("\n\nTENURE VS PROFITABILITY:")
display(tenure_analysis)
```

```python
# Save outputs
customer_clv.to_csv('customer_lifetime_value_analysis.csv', index=False)
unprofitable_repeat.to_csv('unprofitable_repeat_customers.csv', index=False)
```
