```python
# Exclude demographic/metadata columns from analysis
demo_cols = ['latitude', 'longitude', 'altitude', 'precision', 
             'survey_date', 'governorate', 'district', 'village', 'respondent_id']

# Get columns to analyze (exclude demographics)
analysis_cols = [col for col in erms_df.columns if col not in demo_cols]

# Identify open-ended columns (high unique ratio, text responses)
open_ended_cols = []
for col in analysis_cols:
    non_null_count = erms_df[col].notna().sum()
    if non_null_count > 0:
        unique_ratio = erms_df[col].nunique() / non_null_count
        # Check if it's object/string type and has high unique ratio
        if erms_df[col].dtype == 'object':
            # If >50% unique values OR average text length > 20 chars, likely open-ended
            avg_length = erms_df[col].dropna().astype(str).str.len().mean()
            if unique_ratio > 0.5 or avg_length > 20:
                open_ended_cols.append(col)

# Identify binary columns (likely multiple response)
binary_cols = []
for col in analysis_cols:
    if col not in open_ended_cols:  # Skip open-ended
        unique_vals = erms_df[col].dropna().unique()
        # Check if column is binary (0/1 or True/False)
        if len(unique_vals) <= 2 and set(unique_vals).issubset({0, 1, '0', '1', True, False}):
            binary_cols.append(col)

# Group binary columns by common prefixes
from collections import defaultdict
mr_groups = defaultdict(list)

for col in binary_cols:
    parts = col.rsplit('_', 1)
    if len(parts) > 1:
        prefix = parts[0]
        mr_groups[prefix].append(col)
    else:
        mr_groups['standalone'].append(col)

# Multiple response columns (binary columns grouped together)
mr_columns = binary_cols

# Single response columns (everything else, excluding open-ended)
sr_columns = [col for col in analysis_cols 
              if col not in binary_cols and col not in open_ended_cols]

# Display results
print("=" * 80)
print("COLUMN CLASSIFICATION ANALYSIS")
print("=" * 80)

print(f"\n🗑️ OPEN-ENDED COLUMNS (TO DROP) ({len(open_ended_cols)} total):")
print("-" * 80)
for col in open_ended_cols:
    unique_ratio = erms_df[col].nunique() / erms_df[col].notna().sum()
    avg_len = erms_df[col].dropna().astype(str).str.len().mean()
    print(f"  - {col}")
    print(f"    (Unique ratio: {unique_ratio:.2%}, Avg length: {avg_len:.0f} chars)")

print(f"\n\n📊 MULTIPLE RESPONSE COLUMNS ({len(mr_columns)} total):")
print("-" * 80)
for prefix, cols in sorted(mr_groups.items()):
    if len(cols) > 1:
        print(f"\n{prefix}:")
        for col in cols:
            print(f"  - {col}")

print(f"\n\n📝 SINGLE RESPONSE COLUMNS ({len(sr_columns)} total):")
print("-" * 80)
for col in sr_columns[:20]:
    unique_count = erms_df[col].nunique()
    print(f"  - {col} ({unique_count} unique values)")
if len(sr_columns) > 20:
    print(f"  ... and {len(sr_columns) - 20} more")

print(f"\n\n✅ SUMMARY:")
print(f"  Open-ended (dropped): {len(open_ended_cols)}")
print(f"  Multiple Response: {len(mr_columns)}")
print(f"  Single Response: {len(sr_columns)}")

# Create the final lists
erms_mr = mr_columns
erms_sr = sr_columns

# Drop open-ended columns from the dataframe
erms_df_clean = erms_df.drop(columns=open_ended_cols)

print(f"\n✓ Created: 'mr_var' and 'sr_var' lists")
print(f"✓ Created: 'erms_df_clean' (without open-ended columns)")
print(f"  Original shape: {erms_df.shape}")
print(f"  Cleaned shape: {erms_df_clean.shape}")
```
